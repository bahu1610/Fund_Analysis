# Fund Analysis with Python

```python
# Importing the necessary libraries
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("Fund.xlsx", header = None)
```

```python
data.head(10)
```

```python
# Let's drop the first column
data.drop(data.columns[0], axis = 1, inplace = True)
data.head()
```

```python
# Let's drop the first 8 rows
data = data.drop(data.index[:9])

#Let's make the ninth row the header
data.columns = data.iloc[0]
data = data.drop(data.index[0])
data.head()
```

```python
data.reset_index(drop=True, inplace=True)
data.head()
```

```python
# Let us remove the time constraint from the date
from datetime import datetime
data.Date = pd.to_datetime(data.Date)

# Now separate time and date
data['Time'] = data['Date'].dt.time
data['Date'] = data['Date'].dt.date
data.drop(columns = "Time", inplace = True)
data.head()
```

```python
data.rename(columns = {"Fund 1" : "F1", "Benchmark": "BR"}, inplace = True)
```

## Formulating the NAV Chart and hypothesis

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, (ax1, ax2) = plt.subplots(2,1, figsize = (10,10))
ax1.plot("Date", "F1", data = data, label = "Fund 1", color = "red")
ax2.plot("Date", "BR", data = data, label = "Benchmark", color = "blue")
ax1.legend()
ax2.legend()
plt.show()
```

## Total Return

```python
# First: Let us calculate the total accumuated return from NAV
TR_Fund1 = ((data["F1"].iloc[-1] - data["F1"].iloc[0])/data["F1"].iloc[0])
TR_Benchmark = ((data["BR"].iloc[-1] - data["BR"].iloc[0])/data["BR"].iloc[0])

print(f"The total return from Fund 1 is {round(TR_Fund1*100,3)}%")
print(f"The total return from Benchmark Fund is {round(TR_Benchmark*100,3)}%")
```

## Daily Return of the Fund

```python
# Next let us look at Fund 1 daily return as well as Benchmark's daily return

data["F1_Daily"] = data["F1"].pct_change()
data["BR_Daily"] = data["BR"].pct_change()
F1_ADR = data["F1_Daily"].mean()
BR_ADR = data["BR_Daily"].mean()

print(f"The average daily return from Fund 1 is {round(F1_ADR*100,3)}%")
print(f"The average daily return from Benchmark Fund is {round(BR_ADR*100,3)}%")
```

## Formulating Daily Returns Chart for both funds

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize = (10,10))
ax1.hist("F1_Daily", data = data, label = "Fund 1", density = True, color = "red", histtype = "step")
ax2.hist("BR_Daily", data = data, label = "Benchmark",  color = "blue", histtype = "step")
# How to add a curve line of normal probability
data["F1_Daily"].plot.kde(ax = ax1, label = "Fund 1", color = "black")
data["BR_Daily"].plot.kde(ax = ax2, label = "Benchmark", color = "black")
ax1.legend()
ax2.legend()
plt.show()

# Calculating skewness and kurtosis of both
print(f"The skewness of Fund 1 is {round(data['F1_Daily'].skew(),3)}")
print(f"The skewness of Benchmark Fund is {round(data['BR_Daily'].skew(),3)}")
print(f"The kurtosis of Fund 1 is {round(data['F1_Daily'].kurtosis(),3)}")
print(f"The kurtosis of Benchmark Fund is {round(data['BR_Daily'].kurtosis(),3)}")
```

## Annualized Return

```python
# Let's calculate cumulative return to get annualized return of daily return percentage change from F1
# Annualized Return = (1 + Cumulative Return)^365/n

data["F1_Cum_Return"] = (1 + data["F1_Daily"]).cumprod()
data["BR_Cum_Return"] = (1 + data["BR_Daily"]).cumprod()

n = data["Date"].iloc[-1] - data["Date"].iloc[0]
n = n.days

AR_F1 = ((data["F1_Cum_Return"].iloc[-1])**(365/n) - 1)
AR_BR = ((data["BR_Cum_Return"].iloc[-1])**(365/n) - 1)

print(f"The annualized return from Fund 1 is {round(AR_F1*100,3)}%")
print(f"The annualized return from Benchmark Fund is {round(AR_BR*100,3)}%")

labels = ["Fund 1", "Benchmark"]
plt.figure(figsize = (5,5))

plt.bar(labels, [AR_F1, AR_BR], color = ["red", "blue"])
plt.ylim(-0.05, 0)
plt.ylabel("Annualized Return")
plt.title("Annualized Return (%)")
plt.show()


```

## Descriptive Statistics

```python
# Correlation between fund 1 & benchmark

Correl = round(data["F1_Daily"].corr(data["BR_Daily"]),3)
print(f"The correlation between Fund 1 and Benchmark is {Correl}")

F1_Std = data["F1_Daily"].std()
BR_Std = data["BR_Daily"].std()

print(f"The standard deviation of Fund 1 is {round(F1_Std*100,3)}%")
print(f"The standard deviation of Benchmark Fund is {round(BR_Std*100,3)}%")

Cov = data["F1_Daily"].cov(data["BR_Daily"])
print(f"The covariance between Fund 1 and Benchmark is {round(Cov*100,5)}%")

Var_BR = data["BR_Daily"].var()
print(f"The variance of Benchmark Fund is {round(Var_BR*100,5)}%")

Beta = (Cov/(Var_BR)) # Calcualtion with population variance
print(f"The beta of Fund 1 is {round(Beta,5)}")
```

## APLHA of the Fund

```python
risk_free = 0.03800 # (7:55 AM EDT 09/24/24)
alpha = TR_Fund1 - (risk_free + Beta*(TR_Benchmark - risk_free))
print(f"The alpha of Fund 1 is {round(alpha,3)}")
```

## Sharpe Ratio

```python
# Calculating the Sharpe Ratio

Sharpe = (TR_Fund1 - risk_free)/F1_Std
print(f"The Sharpe Ratio of Fund 1 is {round(Sharpe,3)}")

Sharpe_BR = (TR_Benchmark - risk_free)/BR_Std
print(f"The Sharpe Ratio of Benchmark Fund is {round(Sharpe_BR,3)}")
```

## Downside Risk

```python
data["F1_Peak"] = data["F1_Cum_Return"].cummax()
data['Drawdown'] = (data['F1_Cum_Return'] - data['F1_Peak']) / data['F1_Peak']
plt.figure(figsize=(12, 6))
plt.plot(data.Date, data['Drawdown'], color='red', label='Drawdown')
plt.fill_between(data.Date, data['Drawdown'], color='red', alpha=0.3)
plt.title('Drawdown Over Time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Drawdown', fontsize=14)

plt.legend()
plt.grid()
plt.show()

# Getting maximum drawdown

max_drawdown = data['Drawdown'].min()
print(f"The maximum drawdown of Fund 1 is {round(max_drawdown*100,3)}%")
```
